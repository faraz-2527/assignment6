```python
#1. Write a Python program to construct the following pattern, using a nested for loop.

num_lines = 5


for i in range(1, num_lines + 1):
    for j in range(i):
        print('*', end=' ')
    print() 


for i in range(num_lines - 1, 0, -1):
    for j in range(i):
        print('*', end=' ')
    print()  
```

```python
#2. Write a Python program to check whether an alphabet which is given by user in input prompt is a vowel or consonant.

def check_alphabet(char):
    vowels = 'aeiouAEIOU'
    if char in vowels:
        return 'vowel'
    else:
        return 'consonant'

def main():
    try:
        user_input = "a" 
        user_input = user_input.strip()  
        
        if len(user_input) != 1 or not user_input.isalpha():
            print("Invalid input. Please enter a single alphabetic character.")
        else:
            result = check_alphabet(user_input)
            print(f"The character '{user_input}' is a {result}.")
    
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
#3. Write a Python program to create the multiplication table (from 1 to 10)
#of a number and the number will be given by user in input prompt.
#Expected Output:
def generate_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")

def main():
    try:
        user_input = "7" 
        number = int(user_input)
        generate_multiplication_table(number)
    except ValueError:
        print("Invalid input. Please enter a valid integer.")

if __name__ == "__main__":
    main()
```

```python
#4. Write a Python program to guess a number between 1 and 9.Note : User is prompted to enter a guess. If the user guesses wrong thenthe prompt appears again until the guess is correct, on successful guess,user will get a "Well guessed!" message, and the program will exit.import random

def guess_number():
    number_to_guess = random.randint(1, 9)
    predefined_guesses = [3, 5, 7]  
    for guess in predefined_guesses:
        print(f"Guess: {guess}")
        if guess == number_to_guess:
            print("Well guessed!")
            return
    print("No correct guesses in predefined set.")

guess_number()
```

```python
# 5. Write a  Python program to filter the height and width of students, which are stored in a dictionary.Original Dictionary:{'Cierra Vega': (6.2, 70), 'Alden Cantrell': (5.9, 65), 'Kierra Gentry': (6.0, 68), 'Pierre Cox': (5.8, 66)}Height > 6ft and Weight> 70kg:{'Cierra Vega': (6.2, 70)}

students = {
    'Cierra Vega': (6.2, 70),
    'Alden Cantrell': (5.9, 65),
    'Kierra Gentry': (6.0, 68),
    'Pierre Cox': (5.8, 66)
}

filtered_students = {name: (height, weight) for name, (height, weight) in students.items() if height > 6 and weight > 70}

print("Original Dictionary:")
print(students)
print("Height > 6ft and Weight > 70kg:")
print(filtered_students)
```

```python
#6. A  Python dictionary contains List as a value. Write a Python program to clear the list values in the said dictionary.Original Dictionary:{'C1': [10, 20, 30], 'C2': [20, 30, 40], 'C3': [12, 34]}Clear the list values in the said dictionary:{'C1': [], 'C2': [], 'C3': []}

original_dict = {
    'C1': [10, 20, 30],
    'C2': [20, 30, 40],
    'C3': [12, 34]
}

for key in original_dict:
    original_dict[key] = []

print("Original Dictionary:")
print(original_dict)
print("Clear the list values in the said dictionary:")
print(original_dict)
```

```python
#7 
#7.  Rivers: Make a dictionary containing three major rivers and the countryeach river runs through. One key-value pair might be 'nile': 'egypt'.
#• Use a loop to print a sentence about each river, such as The Nile runsthrough Egypt.
#• Use a loop to print the name of each river included in the dictionary.
#• Use a loop to print the name of each country included in the dictionary.
rivers = {
    'Nile': 'Egypt',
    'Amazon': 'Brazil',
    'Yangtze': 'China'
}

for river, country in rivers.items():
    print(f"The {river} runs through {country}.")

print("\nRivers included in the dictionary:")
for river in rivers.keys():
    print(river)

print("\nCountries included in the dictionary:")
for country in rivers.values():
    print(country)
```
